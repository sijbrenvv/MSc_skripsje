```python
#input_file = 'data/english-results-data.csv'
#df = pd.read_csv(input_file, sep=';')
#df = df.loc[df['WAB Type'] == 'Broca'].iloc[:,[0, 3]]
#df.shape
#!pwd

def get_par_speech(contents: list, test_df, file_name):

    describing = ''
    context = ''
    check = False
    for item in contents:
        if item.startswith('@G:'):
            describing = item.split('\t')[-1]
            continue
        if item.startswith('*INV:'):
            if check:
                context += ' ' + item.split('\t')[-1]
                continue
            else:
                context = item.split('\t')[-1]
                check = True
                continue
        if item.startswith('*PAR:'):
            text = item.split('\t')[-1]
            check = False
        
        test_df.loc[len(test_df)] = [file_name, text, context, describing]
# input_file.split('/')[-1][:-4]
    #return test_df


test_df = pd.DataFrame(columns=['Participant ID', 'PAR', 'Context INV', 'Describing'])

for _, file in broca_df.iterrows():

    if file['Participant ID'].startswith('NEURAL'):
        with open(f"data/Aphasia/all_files/{file['Participant ID'][6:]}.cha", 'r') as f:
            json_content = f.read()
    elif file['Participant ID'].startswith('UMD-'):
        try:
            with open(f"data/Aphasia/all_files/{file['Participant ID'][4:]}.cha", 'r') as f:
                json_content = f.read()
        except FileNotFoundError:
            print(f"This file: (UMD-){file['Participant ID'][-6:]}.cha is not present in the downloaded data",file=sys.stderr)
    elif file['Participant ID'].startswith('fridriksson-'):
        try:
            with open(f"data/Aphasia/all_files/{file['Participant ID'][-6:]}.cha", 'r') as f:
                json_content = f.read()
        except FileNotFoundError:
            print(f"This file: (fridriksson-[1|2]){file['Participant ID'][-6:]}.cha is not present in the downloaded data",file=sys.stderr)
    else:
        with open(f"data/Aphasia/all_files/{file['Participant ID']}.cha", 'r') as f:
            json_content = f.read()
        contents = json_content.split('\n')
        contents = [item for item in contents if item.split('\t')[0] in ['@G:', '*INV:', '*PAR:']]
        get_par_speech(contents, test_df, f"{file['Participant ID']}")


#input_file = 'data/Aphasia/all_files/ACWT01a.cha'
#with open(input_file, 'r') as f:
#    json_content = f.read()

#print(json_content[:1050])
#print(json_content.split('\n'))

#contents = json_content.split('\n')
#print(f"Len contents: {len(contents)}")
#print(f"First 10 contents: {contents[:10]}")

#contents = [item for item in contents if item.split('\t')[0] in ['@G:', '*INV:', '*PAR:']]

#print(f"Len cleaned contents: {len(contents)}")
#print(f"First 10 cleaned contents: {contents[:10]}")


print(f"24 files are missing!!")
print(f"{test_df.shape = }")

```
